```python
import sqlite3
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
con = sqlite3.connect('take.db')
```

```python
tiles = pd.read_sql("select * from cv_piece_raw", con)
```

```python
tiles = tiles.set_index(["episode_id", "id"])
```

```python
tiles.head()
```

```python
referent = pd.read_sql("select * from referent", con)
```

```python
tiles_referent = pd.merge(referent, tiles, how="inner", left_on=["object"], right_on=["id"])
```

```python
tiles_referent
```

```python
words = pd.read_sql("select * from hand", con)
```

```python
words.head()
```

```python
tiles_referent_words = pd.merge(tiles_referent, words, how="inner", on=["episode_id"])
```

```python
rot = tiles_referent_words[tiles_referent_words.word == "rot"]
non_rot = tiles_referent_words[tiles_referent_words.word != "rot"]
```

```python
np.average(rot["r"]), np.average(non_rot["r"])
```

```python
links = tiles_referent_words[tiles_referent_words.word == "links"]
non_links = tiles_referent_words[tiles_referent_words.word != "links"]
```

```python
np.average(links["pos_x"]), np.average(non_links["pos_x"])
```
