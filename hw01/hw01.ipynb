## NOAA Climate Datasets
- What is the description of the data? 

        NCEI's climate and weather datasets, products, and various web pages and resources.
        NCEI is the National Centers for Environmental Information and these datasets come from NOAA, the National 
        Oceanic and Atmospheric Administration.  
    
- What is the data for?

        For use by the public, scientists and emergency responders.
        
- How was the data collected?

        Devices such as land-based stations, satellites, radars, models, weather balloons, ships buoys and platforms, and 
        observing paleoclimate data.
        
- What format is the data in?

        Tabular data in a pdf file format, tab delimited files, excel files, visualiztions, csv files.
        
- Is the data up-to-date/recent?

        Some datasets are current and some are varying/not current.
        
- What kinds of examples are there of the data?
        
        Alot of the datasets can be filtered by geographic region and then plotted on a map. For example, the US climate
        normals dataset allows you to pick a station on a map, then get daily/hourly/monthly data on long term climatic
        averages for that station.
        
- Is the data easily accessible?

        At a quick glance, the data seems fairly accessible. As input into a python program, not so much. I foresee alot
        of parsing required in order to load the data into a python program.
        
- What restrictions are there on using the data?

        Some data are subject to a certification, consultation fee or handling charge.
        
- Can you think of something you could learn from analyzing this data?
        
        I'd be curious about the homelessness rates compared to the average climate of a geographic location in the 
        USA. I've heard some people mention that it is easier for the homeless to survive in more forgiving climates
        like Southern California's, but I'd like to back that up with data and research.

- Can you think of some issues or challenges (e.g., is the data too old? Is data missing?) you might have if you analyzed this data?
        
        As I mentioned above, a lot of parsing and data wrangling will be required for a python program to consume this
        data easily. Only a couple of the datasets seem to be in csv file format which means I'd have to write some code
        to parse a pdf file which is not always the easiest thing.
        
        Also, several of the datasets were not current.

## Websites from CommonCrawl
- What is the description of the data?

        Largest hyperlink graph publically available.

- What is the data for?

        The creators of this data hope that it is used for things like search algorithms, SPAM detection,
        graph analysis, and for web science researchers to study.

- How was the data collected?

        A web crawler that visits webpages all over the internet and scrapes data.
        
- What format is the data in?

        WARC, web archive files. Metadata and textual data is stored WAT and WET files.

- Is the data up-to-date/recent?

        Yes.

- What kinds of examples are there of the data?

        There are examples of the WARC files that show things about the web crawlers activity and the
        responses it received from the web pages.

- Is the data easily accessible?

        Not as easy as downloading a CSV file.You have to create an AWS account and then know how to run some code to 
        retrieve it from the AWS servers. Then, you will need enough bandwidth and storage to download/store the data 
        since the datasets are massive.

- What restrictions are there on using the data?

        Don't use the data to abuse or harass others, invade other's privacy, harm minors, violate IP laws, spam or
        stalk people, impersonate people, etc. Basically, don't use the data for criminal purposes.

- Can you think of something you could learn from analyzing this data?

        I'd like to use it to study viral media and how it becomes popularized. Sometimes there is no sense in why
        a video/picture becomes viral but studying the way it propogates the internet could help.

- Can you think of some issues or challenges (e.g., is the data too old? Is data missing?) you might have if you analyzed this data?

        Biggest challenge that stands out to me is download/storing datasets of this size. Also, doing work on this
        data would be costly and require some powerful hardware.

## UCSD Network Telescope, IPv4 /8 net

- What is the description of the data?

        A continuous view of anomalous, unsolicited internet traffic, or Internet Background Radiation.

- What is the data for?

        Collecting data on DDOS attacks, autoamted spread of worms and viruses, scanner of address space by attackers
        or malware looking for vulnerable targets.

- How was the data collected?

        UCSD set up a globally routed /8 network which makes up approximately 1/256th of all IPV4 addresses. Then, they
        discarded any legitimate network traffic and stored the rest.

- What format is the data in?

        PCAP files.

- Is the data up-to-date/recent?

        Yes, it is updated every day.
    
- What kinds of examples are there of the data?

        There are none. UCSD considers this data highly sensitive and does not release examples.
        
- Is the data easily accessible?

        No, you must submit a request to download the data and create an IMPACT account as well as agree to a
        memorandum. These requests can take several days to weeks to process and can be denied.

- What restrictions are there on using the data?

        You cannot download the raw data at all. The data can only be analyzed on provisioned servers.

- Can you think of something you could learn from analyzing this data?

        I could learn alot about which IP address ranges, ports and applications are most vulnerable. When I first
        created my own cloud server, I noticed I was being sent all kinds of invalid requests to applications/ports
        that didn't even exist on my server. An attacker even got network access to my server once and I had to
        reconfigure the security and eventualy take the server down.

- Can you think of some issues or challenges (e.g., is the data too old? Is data missing?) you might have if you analyzed this data?

        Even getting access to this data seems highly challenging. If I had access to this data, the hardest
        part would be understanding network traces since I'm not the most experienced with that.
