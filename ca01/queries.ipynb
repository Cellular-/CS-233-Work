```python
import pandas as pd
```

```python
employees = pd.read_csv("employees.csv")
employees.columns = [column.replace(" ", "_") for column in employees.columns]
```

```python
above_100k = employees.query("Salary > 100000")
above_100k
```

```python
females = employees.query("Gender == 'Female'")
males = employees.query("Gender == 'Male'")

print("Females:", len(females), "Males:", len(males))
```

```python
employees.dropna

types = set(employees.Team)

for t in types:
    senior_mgmt = employees.query("Senior_Management == True and Team == '{}'".format(t))
    non_senior_mgmt = employees.query("Senior_Management == False and Team == '{}'".format(t))
    
    print("Team: {} Num Senior: {} Num Non-Senior: {}".format(t, len(senior_mgmt), len(non_senior_mgmt)))
```

```python
employees.groupby(["Team", "Senior_Management"]).size()
```
